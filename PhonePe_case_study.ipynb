## Task 5: Data Visualization

### 5.1: Plot the total transactions and amount over time for a selected state


### 1. Create a line plot showing the total number of transactions and the total transaction amountover time (years and quarters) for a selected state.[Hint: you can select any state, maybe your home state or state with max transactions]

```python
state_df = state_data[state_data['State'] == 'Jharkhand']
state_df = state_df.groupby(['Year','Quarter'])[['Transactions','Amount (INR)']].sum().reset_index()
state_df['Time'] = state_df['Year'].astype(str) + ' Q' + state_df['Quarter'].astype(str)

import matplotlib.pyplot as plt

state_df = state_df.sort_values(by=['Year','Quarter'])

fig, ax1 = plt.subplots(figsize=(12,6))

# Line 1 → Transactions
ax1.plot(state_df['Time'], state_df['Transactions'], color='blue', marker='o', label='Transactions')
ax1.set_xlabel('Time (Year-Quarter)')
ax1.set_ylabel('Transactions', color='blue')

# Second axis → Amount
ax2 = ax1.twinx()
ax2.plot(state_df['Time'], state_df['Amount (INR)'], color='red', marker='o', label='Amount (INR)')
ax2.set_ylabel('Amount (INR)', color='red')

# Title
plt.title('Transaction Trends in Jharkhand')

# Rotate x labels
plt.xticks(rotation=90)

# Layout
fig.tight_layout()

plt.show()
```

### 5.2: Create a pie chart showing the distribution of transaction types for a specific quarter

### 1. Create a pie chart showing the distribution of different transaction types for a selected state and Quarter

```python
latest_year = txn_Split['Year'].max()

df1 = txn_Split[
    (txn_Split['State'] == 'Jharkhand') &
    (txn_Split['Year'] == latest_year) &
    (txn_Split['Quarter'] == 1)
]
temp = df1.groupby('Transaction Type')['Transactions'].sum().reset_index()

import matplotlib.pyplot as plt

plt.figure(figsize=(8,8))

plt.pie(
    temp['Transactions'],
    labels=temp['Transaction Type'],
    autopct='%1.1f%%',
    startangle=90

)

plt.title('Transaction Type Distribution - Jharkhand (Q1)')

plt.show()
```
